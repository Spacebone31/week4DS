```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nnqomariyah/Fundamentals_of_Data_Science/main/week_3/data/movies_metadata.csv')
```

```python
# Function to clean and parse the JSON-like columns, safely handling empty values
def clean_json_like_column(col_value):
    if pd.isna(col_value):  # Check for NaN or missing values
        return None
    try:
        # Convert single quotes to double quotes for valid JSON format
        col_value = col_value.replace("'", "\"")
        return json.loads(col_value) if isinstance(col_value, str) and (col_value.startswith('{') or col_value.startswith('[')) else col_value
    except Exception:
        return None  # Return None if parsing fails

# Apply the cleaning function to relevant columns
df['belongs_to_collection'] = df['belongs_to_collection'].apply(clean_json_like_column)
df['genres'] = df['genres'].apply(clean_json_like_column)
df['production_companies'] = df['production_companies'].apply(clean_json_like_column)
df['production_countries'] = df['production_countries'].apply(clean_json_like_column)
df['spoken_languages'] = df['spoken_languages'].apply(clean_json_like_column)

# Extract and separate 'belongs_to_collection' into individual columns
df['collection_id'] = df['belongs_to_collection'].apply(lambda x: x['id'] if isinstance(x, dict) else None)
df['collection_name'] = df['belongs_to_collection'].apply(lambda x: x['name'] if isinstance(x, dict) else None)
df['collection_poster_path'] = df['belongs_to_collection'].apply(lambda x: x['poster_path'] if isinstance(x, dict) else None)
df['collection_backdrop_path'] = df['belongs_to_collection'].apply(lambda x: x['backdrop_path'] if isinstance(x, dict) else None)

# Extract and separate 'genres' into individual genre names
df['genre'] = df['genres'].apply(lambda x: ', '.join([genre['name'] for genre in x]) if isinstance(x, list) else None)

# Extract and separate 'production_companies' into individual company names
df['production_company'] = df['production_companies'].apply(lambda x: ', '.join([company['name'] for company in x]) if isinstance(x, list) else None)

# Extract and separate 'production_countries' into individual country names
df['production_country'] = df['production_countries'].apply(lambda x: ', '.join([country['name'] for country in x]) if isinstance(x, list) else None)

# Extract and separate 'spoken_languages' into individual spoken languages
df['spoken_language'] = df['spoken_languages'].apply(lambda x: ', '.join([lang['name'] for lang in x]) if isinstance(x, list) else None)

# Drop the original nested columns if needed
df_cleaned = df.drop(columns=['belongs_to_collection', 'genres', 'production_companies', 'production_countries', 'spoken_languages'])

# Display the cleaned DataFrame with separated columns
df_cleaned
```
